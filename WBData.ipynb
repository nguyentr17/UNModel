# Importing data from World Bank


```python
import pandas as pd
from pandas.io import wb
import wbdata
#reference : http://nbviewer.jupyter.org/github/psychemedia/dataviz4development/blob/master/dataloaders/World%20Bank%20Demo.ipynb


#Get a full list of indicators into a pandas dataframe
df=wb.get_indicators()

#Delete columns we don't need
del df['source']
del df['sourceNote']
del df['sourceOrganization']
del df['topics']

#Tables return upon search keyword
# General format: df[df['name'].str.contains("- key word search-")]
df = df[df['name'].str.contains("GDP per capita")]

#Build a dataframe of selected indicators
selected = pd.DataFrame()

#Add each row to a new dataframe upon selection. 
# General format :  new_row = df.loc[[-index of selected-]]
new_row = df.loc[[7439]]
selected = selected.append(new_row)
selected

#Build dictionary of selected indicator to feed in wbdata.get_dataframe 
indicator_dict = dict(zip(selected.id, selected.name))


#data ready to be analyzed (with all countries)
data = wbdata.get_dataframe(indicator_dict, data_date=None, convert_date=False, keep_levels=False)
data = data.reset_index()
data
```
